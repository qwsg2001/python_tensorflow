```python
model = load_model('dino.h5')
```

```python
template9 = cv.imread('images/2_Bird_Down.png', 0)
template10 = cv.imread('images/2_Bird_Up.png', 0)
```

```python
for i in range(1, 17):
    locals()["w" + str(i)], locals()["h" + str(i)] = locals()["template" + str(i)].shape[::-1]
```

```python
def process_img(original_image):

        img_gray = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

        return img_gray
```

```python
while(True):
        data = [0, 0, 0, 0, 0, 0, 0]
        Input =[0, 0, 0, 0, 0, 0]

        screen =  np.array(ImageGrab.grab(bbox=(0, 0, 800, 800)))
        img_gray = process_img(screen)

        for i in range(1, 17):
            locals()["res" + str(i)] = cv.matchTemplate(img_gray, locals()["template" + str(i)], cv.TM_CCOEFF_NORMED)

        threshold = 0.8
        for i in range(1, 17):
            locals()["loc" + str(i)] = np.where(locals()["res" + str(i)] >= threshold)
        
        
        #나무그림분류
        for pt in zip(*loc3[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w3, pt[1] + h3), (0, 0, 255), 2)
                data[0] = pt[0]
                break
        
        for pt in zip(*loc4[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w4, pt[1] + h4), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc5[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w5, pt[1] + h5), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc6[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w6, pt[1] + h6), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc7[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w7, pt[1] + h7), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc8[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w8, pt[1] + h8), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc11[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w11, pt[1] + h11), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc12[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w12, pt[1] + h12), (0, 0, 255), 2)
                data[0] = pt[0]
                break
        
        for pt in zip(*loc13[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w13, pt[1] + h13), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc14[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w14, pt[1] + h14), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc15[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w15, pt[1] + h15), (0, 0, 255), 2)
                data[0] = pt[0]
                break
                
        for pt in zip(*loc16[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w16, pt[1] + h16), (0, 0, 255), 2)
                data[0] = pt[0]
                break
        
        #새그림 분류
        for pt in zip(*loc1[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w1, pt[1] + h1), (0, 0, 255), 2)
                data[2] = pt[0]
                data[3] = pt[1]
                break

        for pt in zip(*loc2[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w2, pt[1] + h2), (0, 0, 255), 2)
                data[4] = pt[0]
                data[5] = pt[1]
                break
        
        for pt in zip(*loc9[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w9, pt[1] + h9), (0, 0, 255), 2)
                data[2] = pt[0]
                data[3] = pt[1]
                break

        for pt in zip(*loc10[::-1]):
                cv.rectangle(img_gray, pt, (pt[0] + w10, pt[1] + h10), (0, 0, 255), 2)
                data[4] = pt[0]
                data[5] = pt[1]
                break

        Input[0] = data[0]
        Input[1] = data[1]
        Input[2] = data[2]
        Input[3] = data[3]
        Input[4] = data[4]
        Input[5] = data[5]

        result = model.predict(np.expand_dims(Input, axis=0))

        if result > 0.223:
            keyboard.press('up')
        else:
            keyboard.release('up')
    
        cv2.imshow('window', img_gray)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            break
```
